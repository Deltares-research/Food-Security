```python
dfs = []
for file in src_dir.glob("water*.csv"):
    case = file.stem.split("-")[-1]
    df = pd.read_csv(file, index_col=0)
    df["case"] = case
    dfs.append(df)

water_df = pd.concat(dfs).reset_index(drop=True)
water_df.to_csv(src_dir / "water-productivity-50-years-combined.csv")
```

```python
dfs = []
for file in src_dir.glob("corrected*.csv"):
    case = file.stem.split("-")[-1]
    df = pd.read_csv(file, index_col=0)
    df["case"] = case
    dfs.append(df)

yield_df = pd.concat(dfs).reset_index(drop=True)
yield_df.to_csv(src_dir / "corrected-yield-combined.csv")
```
